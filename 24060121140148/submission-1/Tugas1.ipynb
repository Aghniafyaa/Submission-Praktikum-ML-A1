Import Libaries

```python
#Untuk mengimpor pustaka Pandas ke dalam kode Python.
import pandas as pd
```

```python
#Variabel url digunakan untuk menyimpan URL tempat dataset breast cancer tersedia.
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer/breast-cancer.data"

#Variabel names berisi daftar nama kolom yang akan digunakan untuk DataFrame yang akan dibaca.
names = ['class', 'age', 'menopause', 'tumor-size', 'inv-nodes', 'node-caps', 'deg-malig', 'breast', 'breast-quad', 'irradiat']

#Untuk membaca dataset dari URL yang diberikan dan memuatnya ke dalam sebuah DataFrame menggunakan pustaka Pandas.
dataset = pd.read_csv(url, names=names)
```

```python
#Melihat dimensi dataset
dataset.shape
```

dataset.shape menghasilkan sebuah tuple yang berisi dua angka, di mana angka pertama adalah jumlah baris (instances) dalam DataFrame, dan angka kedua adalah jumlah kolom (atribut). Ini berarti DataFrame memiliki 286 baris dan 10 kolom.

```python
#Ringkasan Statistik
(dataset.describe())
```

Hasilnya adalah tabel yang memuat beberapa statistik deskriptif penting mengenai kolom-kolom dalam dataset. Digunakan untuk memberikan pandangan cepat tentang distribusi dan karakteristik data dalam setiap kolom dataset.

```python
#Ringkasan Statistik
import matplotlib.pyplot as plt
dataset.plot(kind='box', subplots=True, layout=(2,2), sharex=False, sharey=False)
plt.show()
```

Hasil visualisasi box plot menunnjukkan statistik penting. box plot diatas menunjukkan bahwa letak garis median tida berada  di tengah kotak artinya data miring (tidak simetris)

```python
# Visualisasi Data Menggunakan Plot Univariat
dataset.hist()
plt.show()
```

 menghasilkan sejumlah plot histogram terpisah, masing-masing menggambarkan distribusi data dalam kolom-kolom dataset.

```python
#Visualisasi Data Menggunakan Plot Multivariat
import pandas
from pandas.plotting import scatter_matrix
scatter_matrix(dataset)
plt.show()
```

Scatter matrix adalah cara yang berguna untuk memvisualisasikan hubungan antara semua pasangan atribut dalam dataset. Hasilnya akan menghasilkan matriks dari scatter plot yang menunjukkan sebaran data antara semua pasangan atribut dalam dataset.
